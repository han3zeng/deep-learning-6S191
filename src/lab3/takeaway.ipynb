## Tokenization
### Overview
Break sentences into sequence of chunks (atomic-text-unit) and map each chunk into numerical token whicih should be represented by a vector

### Types (3 different based)
- word
    - Need large vocabulary
    - Prboelm: Can't understand the unseen word
- char
    - Extreme long sequence
    - Lose word level meaning
- subword (common practice) (Byte-pair encoding)
    - Break words into subwords based on frequency so it mixes both word and char level.
    - Merge the most frequent pair together as single unit
